# Part 1 - scrapping Wiki for data

```python
#Scrapping Wikipedia page with native pandas function
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)
```

```python
#viewing the scrapped table
print(dfs[0])
```

```python
#creating a new dataframe called Toronto to store the content of the scrapped table
toronto = dfs[0]
```

```python
#checking the new toronto df
toronto
```

```python
#removing rows where Borough = Not Assigned
toronto=toronto[toronto.Borough != "Not assigned"]
```

```python
#viewing the new toronto df without not assigned rows
toronto
```

```python
#checking for any rows where Neighbourhood = Not assigned
toronto.loc[toronto['Neighbourhood'] == 'Not assigned']
```

```python
#renaming columns
toronto = toronto.rename(columns={"Postal Code":"PostalCode", "Neighbourhood":"Neighborhood"})
```

```python
#checking the new toronto df
toronto
```

```python
#resetting the index column and removing the previous index
toronto=toronto.reset_index()
toronto = toronto.drop(columns='index')
```

```python
#viewing the new toronto df
toronto
```
